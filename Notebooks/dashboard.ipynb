```python
data = pd.read_csv("data_dash.csv")
data.head()
```

# 1. Un mapa que muestre la cantidad de hospitalizados debido al COVID-19 por Estado.

```python
df = data.groupby(['code', 'Year', 'Mes'])[['hosp_covid']].sum()
df.reset_index(inplace=True)
```

```python
option1 = 2020
df_2020 = df.copy()
df_2020 = df_2020[df_2020["Year"] == option1]


fig = px.choropleth(
        data_frame=df_2020,
        locationmode='USA-states',
        locations='code',
        scope="usa",
        color='hosp_covid',
        hover_data=['code', 'hosp_covid'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        labels={'Hospitalizados por COVID'},
        template='plotly_dark')


fig.update_layout( title_text = 'Hospitalizados por COVID-19 en el  2020')

fig.show()
```

```python
option2 = 2021
df_2021 = df.copy()
df_2021 = df_2021[df_2021["Year"] == option2]


fig = px.choropleth(
        data_frame=df_2021,
        locationmode='USA-states',
        locations='code',
        scope="usa",
        color='hosp_covid',
        hover_data=['code', 'hosp_covid'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        labels={'Hospitalizados por COVID'},
        template='plotly_dark'
)

fig.update_layout( title_text = 'Hospitalizados por COVID-19 en el  2021')

fig.show()
```

```python
option3 = 2022
df_2022 = df.copy()
df_2022 = df_2022[df_2022["Year"] == option2]


fig = px.choropleth(
        data_frame=df_2022,
        locationmode='USA-states',
        locations='code',
        scope="usa",
        color='hosp_covid',
        hover_data=['code', 'hosp_covid'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        labels={'Hospitalizados por COVID'},
        template='plotly_dark')

fig.update_layout( title_text = 'Hospitalizados por COVID-19 en el  2022')

fig.show()
```

# 2. Uso de camas UCI por Estado.

```python
uci = 'uso_uci'
```

```python
year1 = 2020

df1 = data[data['Year'] == year1].sort_values(uci)

px.bar(df1, x = 'code', y = uci, title =' - '.join([uci, str(year1)]), template="seaborn")
```

```python
year2 = 2021

df2 = data[data['Year'] == year2].sort_values(uci)

px.bar(df2, x = 'code', y = uci, title =' - '.join([uci, str(year2)]), template="seaborn")
```

```python
year3 = 2022

df3 = data[data['Year'] == year3].sort_values(uci)

px.bar(df3, x = 'code', y = uci, title =' - '.join([uci, str(year3)]))
```
